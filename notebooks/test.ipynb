```python
import torch
import torch.nn as nn
```

```python
n_vocab = 64
hidden_channels = 192
emb = nn.Embedding(n_vocab, hidden_channels)
```

```python
xs = torch.LongTensor([[1,2,3,1,5,16,7,1], [5,6,7,1,4,9,0,1]])
xs.shape
```

```python
x = emb(xs).transpose(1,2)
print(x.shape)
```

```python
def sequence_mask(length, max_length=None):
    """torch.arange(max_length).unsqueeze(0) < length.unsqueeze(1)"""
    if max_length is None: max_length = length.max()
    x = torch.arange(max_length)
    return x.unsqueeze(0) < length.unsqueeze(1)

mask = sequence_mask(torch.LongTensor([6,7]), max_length=8).unsqueeze(1)
print(mask.shape)  # torch.Size([2, 1, 8])
mask
```
